# Feature selection technics :

### 1. Droping the constant features :
##### In this step we will be removing the features which have constant features which are actually not important for solving the problem statement

```python
import pandas as pd 

data = pd.DataFrame({'A':[1,2,4,8,9],
                   'B':[4,8,7,6,5],
                   'C':[0,0,0,0,0],
                   'D':[1,1,1,1,1]})
```

```python
data.head()
```

#### Variance Threshold
Feature selector that removes all low-variance features (very less number of unique values).

This feature selection algorithm looks only at the features (X), 
not the desired outputs (y), and can thus be used for unsupervised learning.

Using a threshold of 0.01 would mean dropping the column where 99% of the values are similar.

```python
# It will zero variance features

from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
var_thres.get_support()
```

1. True = Need keep, 
2. False = Going to drop

```python
data.columns[var_thres.get_support()]
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
# Loop comprehension, Where we trying to extract those colums which are constant. 
constant_colums = [column for column in data.columns 
                   if column not in data.columns[var_thres.get_support()]]
print(len(constant_colums))
```

### Sigma/mu  (Std devation/Means) < 0.5 those columns also we can use to drop. 

```python
# Column C and D is going to drop
for feature in constant_colums:
    print(feature)
```

```python
data.drop(constant_colums, axis=1)
```
